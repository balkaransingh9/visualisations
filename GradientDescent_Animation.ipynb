```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```
def f(x, y):
    #return (np.sin(x)*y + np.sin(y)*x)
    return (x**2)+(y**2)

x = np.linspace(-6, 6, 60)
y = np.linspace(-6, 6, 60)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(X, Y, Z, cmap='viridis')
ax.scatter(0,0,0, color='red')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(40, 25)
plt.rcParams["figure.figsize"] = (15,10)
```

```
plt.contour(X, Y, Z,color='black')
plt.show()
```

```
# make it flexible so it takes in any number of variables..
import torch

init = [[2.0],[3.5]]
lr = 0.1
iterations = 0
iterates = list()

while iterations < 20:
  x = torch.tensor(init, requires_grad=True)
  y = torch.sum(x[0]**2 + 4*x[1]**2 - 4*x[1])
  y.backward()
  gradient = x.grad.detach()
  ans = []
  for i in range(0,gradient.shape[0]):
    x = init[i][0]-lr*gradient[i][0]
    ans.append([x])
  init = ans
  print(ans)
  iterations += 1
```
