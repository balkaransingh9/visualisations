```
import matplotlib.pyplot as plt 
import matplotlib.animation as animation 
import matplotlib.gridspec as gridspec
import numpy as np 
from IPython.display import HTML
plt.style.use('dark_background')

xval = 0

gs = gridspec.GridSpec(2, 2)
fig = plt.figure(figsize=(16, 8))
ax1 = fig.add_subplot(gs[:, 0]) # row 0, col 0
line, = ax1.plot([], [], lw=2)
triangle, = ax1.plot([], [], lw=2)
dot, = ax1.plot([],[],'wo')
ax1.set_xlim(-1.2,1.2)
ax1.set_ylim(-1.2,1.2)
ax1.axhline(0,linewidth=0.4)
ax1.axvline(0,linewidth=0.4)
ax1.grid(linestyle='--',linewidth=0.3)

ax2 = fig.add_subplot(gs[0, 1]) # row 0, col 1
line2, = ax2.plot([], [], lw=2)
dot2, = ax2.plot([],[],'wo')
ax2.set_xlim(0,10)
ax2.set_ylim(-1.2,1.2)
ax2.axhline(0,linewidth=0.4)
ax2.grid(linestyle='--',linewidth=0.3)


ax3 = fig.add_subplot(gs[1, 1]) # row 1, span all columns
line3, = ax3.plot([], [], lw=2)
dot3, = ax3.plot([],[],'wo')
ax3.set_xlim(0,10)
ax3.set_ylim(-1.2,1.2)
ax3.axhline(0,linewidth=0.4)
ax3.grid(linestyle='--',linewidth=0.3)


def init(): 
	line.set_data([], []) 
	return line, 

xdata, ydata, theta = [], [] , []

def animate(i): 
	global xval
	x = np.cos(xval)
	y = np.sin(xval)
	xdata.append(x)
	theta.append(xval)
	ydata.append(y)
	xval = xval + 0.01
	triData = [[0,x,x,0],[0,0,y,0]]
	triangle.set_data(triData[0],triData[1])
	line.set_data(xdata, ydata)
	dot.set_data(x,y)
	line2.set_data(theta, xdata)
	dot2.set_data(xval,x)
	line3.set_data(theta,ydata)
	dot3.set_data(xval,y)
 
	return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=1300, interval=7, blit=True) 
```
